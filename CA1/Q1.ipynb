ساخت تابع نمونه برداری

```python
def randbin(n,m,p):
    x=np.random.choice(np.array([1,0]),(m,n),p=np.array([p, 1-p]))
    return np.sum(x,1)
```

```python
print(np.random.choice([0,1], size=(2,5), p=[0.1, 0.9]))
```

```python
#practical exp
z=np.array([])
for p in range (0,100):
    a=randbin(500,5000,p/100)
    z = np.append(z, np.average(a))
print(np.average(z))

# theory exp
sum=0
for p in range (0,100):
    sum+=500*(p/100)
exp=sum/100
print(exp)
```

```python
# practical var
k=np.array([])
for p in range (0,100):
    b=randbin(500,5000,p/100)
    for i in b:
        gashtavar=np.array([])
        gashtavar=np.append(gashtavar,(b-np.average(b))*(b-np.average(b)))
    k=np.append(k,np.average(gashtavar))
print(np.average(k))

# theory var
sumvar=0
for p in range (0,100):
    sumvar+=500*(p/100)*(1-(p/100))
var=sumvar/100
print(var)
```
